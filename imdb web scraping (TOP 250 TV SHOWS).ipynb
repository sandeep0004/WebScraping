```python
import pandas as pd
import requests
import bs4
import re
```

```python
url=requests.get("https://www.imdb.com/chart/toptv/?ref_=nv_td_tv250")
data=bs4.BeautifulSoup(url.text,"html.parser")
data
```

```python
Name=[]
Year=[]
Rating=[]
User_Rating=[]
```

```python
name=data.find_all("td",class_="titleColumn")
for names in name:
    names=names.a
    Name.append(names.text)
```

```python
year=data.find_all("td",class_="titleColumn")
for years in year:
    years=years.span
    Year.append(years.text)
```

```python
rating=data.find_all("td",class_="ratingColumn imdbRating")
for ratings in rating:
    ratings=ratings.strong
    Rating.append(ratings.text)
```

```python
user=data.find_all("td",class_="ratingColumn imdbRating")
user
```

```python
for ratings in user:
    ratings=data.strong.get("title")
    ratings=ratings[13:21]
    User_Rating.append(ratings) 
```

```python
dataset=pd.DataFrame({"Name":Name,"Year":Year,"Ratings":Rating,"Users":User_Rating})
```

```python
dataset
```

```python
dataset["Year"]=dataset["Year"].str.replace("(",'',regex=True).str.replace(")",'',regex=True)
```
